```python
import numpy as np
from scipy.linalg import toeplitz, cho_factor, cho_solve
from scipy.optimize import minimize_scalar


def _B_matrix(n_low: int, s: int, agg: str) -> np.ndarray:
    """
    Aggregation matrix B mapping high-freq y (T=n_low*s) to low-freq y (n_low).
    agg:
      - "sum"  : sum of s periods
      - "mean" : average of s periods
      - "last" : last period in the block (e.g., Q4 for annual stock)
      - "first": first period in the block
    """
    T = n_low * s
    B = np.zeros((n_low, T))
    for i in range(n_low):
        j0 = i * s
        if agg == "sum":
            B[i, j0:j0+s] = 1.0
        elif agg == "mean":
            B[i, j0:j0+s] = 1.0 / s
        elif agg == "last":
            B[i, j0 + (s - 1)] = 1.0
        elif agg == "first":
            B[i, j0] = 1.0
        else:
            raise ValueError("agg must be one of {'sum','mean','last','first'}")
    return B


def chow_lin(y_low, x_high, s=4, agg="mean"):
    """
    Chow–Lin disaggregation with AR(1) residuals.
      y_low:  (n_low,) low-frequency series (e.g., annual log(tfpa))
      x_high: (T,) or (T,k) high-frequency indicators (e.g., [1, log(rgdp), log(pop)])
             where T = n_low * s
      s:      frequency ratio (annual->quarterly: s=4)
      agg:    how annual relates to quarters ("mean" typical; "last" for year-end stocks)
    Returns:
      y_high_hat: (T,) estimated high-frequency series
      rho_hat:    AR(1) parameter
    """
    y_low = np.asarray(y_low, dtype=float).reshape(-1)
    Xh = np.asarray(x_high, dtype=float)
    if Xh.ndim == 1:
        Xh = Xh[:, None]

    n_low = y_low.shape[0]
    T = n_low * s
    if Xh.shape[0] != T:
        raise ValueError(f"x_high must have length T=n_low*s={T}, got {Xh.shape[0]}")

    B = _B_matrix(n_low, s, agg)

    def Sigma_ar1(rho):
        idx = np.arange(T)
        c = rho ** idx          # 1D first column of Toeplitz
        return toeplitz(c)      # Toeplitz(c, c)

    def neg_loglike(rho):
        S = Sigma_ar1(rho)
        Omega = B @ S @ B.T
        Xl = B @ Xh

        c, low = cho_factor(Omega, lower=True, check_finite=False)
        Oinv_y = cho_solve((c, low), y_low, check_finite=False)
        Oinv_X = cho_solve((c, low), Xl, check_finite=False)

        beta = np.linalg.solve(Xl.T @ Oinv_X, Xl.T @ Oinv_y)

        e = y_low - Xl @ beta
        Oinv_e = cho_solve((c, low), e, check_finite=False)

        sig2 = (e @ Oinv_e) / n_low
        sign, logdet = np.linalg.slogdet(Omega)
        return 0.5 * (n_low * np.log(sig2) + logdet)

    res = minimize_scalar(neg_loglike, bounds=(-0.99, 0.99), method="bounded",
                          options={"maxiter": 50})
    rho_hat = float(res.x)

    # Final BLUP
    S = Sigma_ar1(rho_hat)
    Omega = B @ S @ B.T
    Xl = B @ Xh

    c, low = cho_factor(Omega, lower=True, check_finite=False)
    beta = np.linalg.solve(
        Xl.T @ cho_solve((c, low), Xl, check_finite=False),
        Xl.T @ cho_solve((c, low), y_low, check_finite=False),
    )

    y_hat_h = Xh @ beta + S @ B.T @ cho_solve(
        (c, low), (y_low - Xl @ beta), check_finite=False
    )

    return y_hat_h.reshape(-1), rho_hat
```

```python
def companion_eigs(A_draw):
    """
    A_draw: (p, K, K) with VAR form y_t = sum_{l=1}^p A_l y_{t-l} + ...
    Returns eigenvalues of the companion matrix.
    """
    p, K, _ = A_draw.shape
    C = np.zeros((K*p, K*p))
    C[:K, :K*p] = np.hstack([A_draw[l] for l in range(p)])
    if p > 1:
        C[K:, :-K] = np.eye(K*(p-1))
    return np.linalg.eigvals(C)

def irf_from_A(A_draw, shock_idx, horizons=40, shock_size=1.0):
    """
    A_draw: (p, K, K)
    shock_idx: index of shocked variable in vars_endog ordering
    horizons: number of steps ahead
    shock_size: size of shock applied to y_t at impact (in the same units as the standardized data)
    Returns: (horizons+1, K) IRF path.
    """
    p, K, _ = A_draw.shape

    # Companion matrix
    C = np.zeros((K*p, K*p))
    C[:K, :K*p] = np.hstack([A_draw[l] for l in range(p)])
    if p > 1:
        C[K:, :-K] = np.eye(K*(p-1))

    # Impact shock to current period
    state = np.zeros(K*p)
    state[shock_idx] = shock_size

    out = np.zeros((horizons+1, K))
    out[0] = state[:K]

    for h in range(1, horizons+1):
        state = C @ state
        out[h] = state[:K]

    return out

def companion_eigs(A_draw):
    """
    A_draw: (lags, K, K) with VAR form y_t = sum_{l=1}^p A_l y_{t-l} + ...
    returns eigenvalues of companion matrix
    """
    p, K, _ = A_draw.shape
    C = np.zeros((K*p, K*p))
    C[:K, :K*p] = np.hstack([A_draw[l] for l in range(p)])
    if p > 1:
        C[K:, :-K] = np.eye(K*(p-1))
    return np.linalg.eigvals(C)
```

```python
# 1) Rename first column to "year"
time_col = df_raw_tfpa.columns[0]
df_raw_tfpa = df_raw_tfpa.rename(columns={time_col: "year"}).copy()

# 2) Coerce year to integer safely (handles strings, 1890.0, etc.)
df_raw_tfpa["year"] = (
    pd.to_numeric(df_raw_tfpa["year"], errors="coerce")
      .astype("Int64")
)

# Drop rows where year couldn't be parsed
df_raw_tfpa = df_raw_tfpa.dropna(subset=["year"])

# 3) Build a DatetimeIndex at Jan 1 of each year
df_raw_tfpa["date"] = pd.to_datetime(df_raw_tfpa["year"].astype(int).astype(str) + "-01-01")

# 4) Set index and sort
df_raw_tfpa = df_raw_tfpa.set_index("date").sort_index()

# 5) Drop helper columns (ignore if missing)
df_raw_tfpa = df_raw_tfpa.drop(columns=["year_frac", "year", "quarter"], errors="ignore")
```

```python
df = df.join(
    df_raw_tfpa[["USA"]].rename(columns={"USA": "tfpa"}),
    how="left"
)
```

```python
df_raw_tfpq = pd.read_excel(data_path2, sheet_name=1, skiprows=1)[:-10]

s = df_raw_tfpq["date"].astype(str)

# "1947:Q1" -> Period('1947Q1', 'Q') -> Timestamp at quarter start
df_raw_tfpq["date_dt"] = (
    pd.PeriodIndex(s.str.replace(":", ""), freq="Q")
      .to_timestamp(how="start")
)

df_raw_tfpq["date"] = (
    pd.PeriodIndex(df_raw_tfpq["date"].astype(str).str.replace(":", ""), freq="Q")
      .to_timestamp(how="start")
)

df_raw_tfpq = df_raw_tfpq.set_index("date")
df_raw_tfpq.index.name = "date"
```

```python
dlog = df_raw_tfpq["dtfp_util"] / 400.0          # quarterly Δlog
logA = dlog.fillna(0).cumsum()          # log level (up to a constant)
A = np.exp(logA - logA.iloc[0])         # level index, starts at 1
df_raw_tfpq["tfpq"] = A
```

```python
df = df.join(df_raw_tfpq["tfpq"], how="left")
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["tfpq"], df["tfpa"], alpha=0.4)
plt.xlabel("tfpq")
plt.ylabel("tfpa")
plt.show()
```

```python
# df: DateTimeIndex already, contains columns tfpa (annual), tfpq (quarterly), rgdp, pop

# 1) Annual tfpa: take the first non-null tfpa per calendar year
ann = (
    df["tfpa"]
    .dropna()
    .groupby(df["tfpa"].dropna().index.year)
    .last()                      # use .first() if that's how your annual rows are stored
)
years = ann.index.astype(int)
y_low = np.log(ann.to_numpy())

# 2) Quarterly index aligned to those annual years (calendar-year quarters)
q_idx = pd.to_datetime([f"{y}-{m:02d}-01" for y in years for m in (1, 4, 7, 10)])

# Pull indicators for exactly those quarters
tmp = df.reindex(q_idx)[["rgdp", "pop"]]

# sanity check: are any quarters missing?
missing = tmp.isna().any(axis=1)
if missing.any():
    raise ValueError(
        "Missing rgdp/pop for some quarters needed by Chow–Lin. "
        f"Example missing dates: {tmp.index[missing][:10].tolist()}"
    )

# Chow–Lin indicator matrix: constant + log(rgdp) + log(pop)
Xh = np.column_stack([
    np.ones(len(q_idx)),
    np.log(tmp["rgdp"].to_numpy()),
    np.log(tmp["pop"].to_numpy()),
])
```

```python
np.isnan(Xh).any(), np.isnan(y_low).any()
```

```python
log_tfp_q, rho = chow_lin(y_low, Xh, s=4, agg="mean")  # try "last" if annual is year-end
tfpa_q = pd.Series(np.exp(log_tfp_q), index=q_idx, name="tfpa_q")
```

```python
df["tfpa_q"] = tfpa_q.reindex(df.index)
```

```python
df[["tfpq","tfpa","tfpa_q"]].interpolate().plot()
```

```python
overlap = df["tfpq"].dropna().index.intersection(df["tfpa_q"].dropna().index)
scale = np.exp(np.nanmedian(np.log(df.loc[overlap, "tfpa_q"]) - np.log(df.loc[overlap, "tfpq"])))

df["tfpq_scaled"] = df["tfpq"] * scale

cut = pd.Timestamp("1947-01-01")
df["tfp_combined"] = df["tfpa_q"]
df.loc[df.index >= cut, "tfp_combined"] = df.loc[df.index >= cut, "tfpq_scaled"]
```

```python
df['rgov'] = df['ngov']/df['pgdp'] #Real gov spending
df['rgov_pc'] = df['rgov']/df['pop'] #Real gov spending per capita
df['rgdp_pc'] = df['rgdp']/df['pop'] #Real gdp per capita
df["def_gdp"] = -1*100*df["ndeficit_nipa"] / df["ngdp"] #deficit to GDP, transform to surplus by * -1
df["debt_gdp"] = 100*df["pubfeddebt_treas"] / df["ngdp"] #debt to GDP
df["news_gdp"] = 100*df["news"] / df["ngdp"] #news to GDP

df["log_rgov"] = np.log(df["rgov"])
df["log_rgov_pc"] = np.log(df["rgov_pc"])
df["log_rgdp"] = np.log(df["rgdp"])
df["log_rgdp_pc"] = np.log(df["rgdp_pc"])
df["log_tfp"] = np.log(df["tfp_combined"])
```

```python
df_ads = pd.read_csv(data_path3)
df_ads["date"] = pd.to_datetime(df_ads["date"])
df_ads = df_ads.set_index("date")

df = df.join(
    df_ads[['def_delta','tbill']],
    how="left"
)
```
